### Federal Election Commission
The US Federal Election Commission publishes data on contributions to political campaigns. This includes contributor names, occupation and employer, address, and contribution amount. The contribution data from the 2012 US presidential election was available as a single 150 megabyte CSV file P00000001-ALL.csv

```python
import pandas as pd
import numpy as np
```

```python
fec_1 = pd.read_csv('datasets/fec/P00000001-ALL_1.csv')
fec = fec_1[fec_1.cand_nm.isin(['Bachmann, Michelle', 'Romney, Mitt'])]
fec

# to read both the files and concatenate into one
# fec_1 = pd.read_csv('datasets/fec/P00000001-ALL_1.csv')
# fec_2 = pd.read_csv('datasets/fec/P00000001-ALL_2.csv')
# fec = pd.concat([fec_1, fec_2])
# fec
```

```python
parties = {"Bachmann, Michelle": "Democrat",
           "Cain, Herman": "Republican",
           "Gingrich, Newt": "Republican",
           "Huntsman, Jon": "Republican",
           "Johnson, Gary Earl": "Republican",
           "McCotter, Thaddeus G": "Republican",
           "Obama, Barack": "Democrat",
           "Paul, Ron": "Republican",
           "Pawlenty, Timothy": "Republican",
           "Perry, Rick": "Republican",
           "Roemer, Charles E. 'Buddy' III": "Republican",
           "Romney, Mitt": "Republican",
           "Santorum, Rick": "Republican"}

fec['party'] = fec.cand_nm.map(parties)
fec
```

```python
# include only contributions and not refund
fec = fec[fec.contb_receipt_amt > 0]
fec
```

```python
# Considering only Obama and Mitt Romney who were the main two candidates

fec_bo_mr = fec[fec['cand_nm'].isin(["Bachmann, Michelle", "Romney, Mitt"])]
fec_bo_mr
```

#### Donation Statistics by Occupation and Employer

```python
fec.contbr_occupation.value_counts()
```

```python
# cleaning up similar occupations by mapping from one occupation to another
occ_mapping = {
   "INFORMATION REQUESTED PER BEST EFFORTS" : "NOT PROVIDED",
   "INFORMATION REQUESTED" : "NOT PROVIDED",
   "INFORMATION REQUESTED (BEST EFFORTS)" : "NOT PROVIDED",
   "C.E.O.": "CEO"
}

def get_occupation(occ):
    return occ_mapping.get(occ, occ)

fec_bo_mr['contbr_occupation'] = fec_bo_mr.contbr_occupation.map(get_occupation)
fec_bo_mr.contbr_occupation
```

```python
# performing the similar cleanup for employees
emp_mapping = {
   "INFORMATION REQUESTED PER BEST EFFORTS" : "NOT PROVIDED",
   "INFORMATION REQUESTED" : "NOT PROVIDED",
   "SELF" : "SELF-EMPLOYED",
   "SELF EMPLOYED" : "SELF-EMPLOYED",
}

def get_employees(emp):
   return emp_mapping.get(emp, emp)

fec_bo_mr['contbr_employer'] = fec_bo_mr.contbr_employer.map(get_employees)

# check and drop duplicates
# fec_bo_mr.duplicated().sum()
# fec_bo_mr = fec_bo_mr.drop_duplicates()

# # check and drop NA
# fec_bo_mr.isnull().sum()
# fec_bo_mr = fec_bo_mr.dropna()

# fec_bo_mr = fec_bo_mr[fec_bo_mr.contb_receipt_amt > 0]
fec_bo_mr
```

```python
# aggregate donation data by party and occupation

by_occupation = fec_bo_mr.pivot_table('contb_receipt_amt', index='contbr_occupation', columns='party', aggfunc=sum)

# donations over 1mn 
over_1mn = by_occupation[by_occupation.sum(axis='columns') > 1000000]
over_1mn
```

```python
# total donations by party for top occupations
over_1mn.plot(kind='barh')
```

```python
def get_top_donation(group, key, n=5):
    total = group.groupby(key)['contb_receipt_amt'].sum()
    return total.nlargest(n)

grouped = fec_bo_mr.groupby('cand_nm')
grouped.apply(get_top_donation, 'contbr_occupation', 10)
```

#### Bucketing Donation Amounts

```python
bins = np.array([0, 1, 10, 100, 1000, 10000, 100_000, 1_000_000, 10_000_000])
labels = pd.cut(fec_bo_mr['contb_receipt_amt'], bins)
labels
```

```python
grouped = fec_bo_mr.groupby(['cand_nm', labels]).size().unstack(level=0)
grouped

# interpretation
# Romney recieved larger small number of donations compared to Michelle
```
